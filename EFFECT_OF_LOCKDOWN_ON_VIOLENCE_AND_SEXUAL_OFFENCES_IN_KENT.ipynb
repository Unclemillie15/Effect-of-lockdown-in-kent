**INTRODUCTION**

The term crime does not have a clear and widely agreed meaning in contemporary criminal law (Lindsay, 2008); however, specific meanings have been given for some purposes under the Trade Union and Labor Relations Act 1992. The most widely held belief is that criminality is a legal category; in other words, anything is a crime if it is declared as such by the appropriate and applicable statute (Lindsay, 2008). According to one suggested description, a crime or offence (or criminal offence) is an act that is detrimental not only to a person but also to a culture, environment, or state ("a public wrong"). Such activities are illegal and punishable by statute (Oxford, 2009; Martin, 2003). In some exceptions, the "act of doing something illegal" (actus reus) must usually be followed by the "will to do something criminal" (mens rea) in order to be regarded as an offence (Martin, 2003).

**EFFECT ON LOCKDOWN ON CRIME IN ENGLAND**

Global officials, as well as foreign and civil society groups, have warned about the pandemic’s potential effect on the occurrence of domestic violence since its inception. The warnings were issued because it was recognised that the pandemic provided conditions conducive to domestic violence: a rise in people’s discomfort and tension (Card and Dahl, 2011); a rise in the amount of time people spent at home; and a variety of negative psychological consequences (Brooks et al., 2020).
Because of drastic shifts in people’s patterns of behaviour during the coronavirus pandemic in the UK, fascinating and, in some cases, dramatic patterns in crime rates and forms of crime have arisen. According to the Guardian (2020), “Criminal activity in certain categories has fallen by as much as 20% in some areas. Crime is evolving as fast as society responds to the new normal”.  Early trends are also starting to emerge in data recently released as a result of studies conducted by organisations such as the ONS. These findings further show that there has been a major fluctuation in the rates and forms of crime perpetrated (as well as the types of crime for which the police are investigating people) since the pandemic era.
At first sight, the data may seem to suggest that, thanks to the lockout and restricted travel of people, the planet seems to be a safer environment. However, the Office of national statistics recorded that “Rape, murder and burglaries fell by roughly a third almost immediately at the start of the lockdown”. Furthermore, except for fraud and electronic abuse, total crime decreased by about 32%, owing primarily to a decrease in domestic robbery and other types of stealing. With large swaths of the population following government advice to ‘Stay at home,’ it is apparent that there have been fewer chances for sexual offences and rape (committed by strangers), and street robberies have fallen dramatically. (Guardian, 2020; Office of National Statistics, 2020).

```python
dates = ['2020-01','2020-02','2020-03', '2020-04', '2020-05', '2020-06', '2020-07', '2020-08', '2020-09', '2020-10',
         '2020-11', '2020-12', '2021-01','2021-02','2021-03']
crimes = ['Violence and sexual offences','Anti-social behaviour','Criminal damage and arson',
          'Shoplifting','Vehicle crime','Public order','Burglary','Drugs','Robbery','Possession of weapons',
          'Bicycle theft','Theft from the person','Other theft','Other crime']
```

```python
def get_data(dates, crimes):
    df_data = []
    for i in dates:
        d_dict = {}
        # df = pd.read_csv("{}-kent-street.csv".format(i))
        df = pd.read_csv("/content/drive/MyDrive/NWP/{}-kent-street.csv".format(i))
        stn = df['Crime type'].value_counts()
        for crime in crimes:
            d_dict[crime] = stn[crime]
        df_data.append(d_dict)
    return df_data
```

```python
#Table of Most commited crimes in each month
##########################################
def get_most_commited_per_month(data):
    d_v = []
    count = 0
    for j in data:
        n_d = {}
        date = dates[count]
        max_v = 0
        max_n = ''
        for i in j:
            if j[i] > max_v:
                max_v, max_n = j[i], i
        n_d['Crime'], n_d['Crime rate'] = max_n, max_v
        d_v.append(n_d)
        count +=1
    return d_v
```

```python
mc_data = get_most_commited_per_month(data)
```

```python
mc_df = pd.DataFrame(mc_data,index=dates)
```

```python
def generate_mc_size(mc_data):
    arr = []
    for i in mc_data:
        arr.append(i['Crime rate'])
    return arr

    return [mc_df.loc[date, "Crime rate"], ]
def generate_mc_labels(dates, mc_data):
    count = 0
    arr = []
    for date in dates:
        label = '{}-{}'.format(date, mc_data[count]['Crime'])
        arr.append(label)
        count+=1
    return arr
```

**EFFECT OF LOCKDOWN ON VOILENCE AND SEXUAL OFFENCES IN UK; KENT**

An increasing body of literature has shown that there is an uptick in abuse and sexual offence cases throughout the UK since the lockdown. Bradbury-Jones and Isham (2020) identify the paradoxical consequences of a pandemic in this way in their latest study. A new study in The Guardian (2020) focuses on an increase in domestic abuse and emphasises a global trend.

This follows a 25% surge in calls relating to domestic abuse incidents in the United Kingdom within a week of stringent social distancing and lockout measures being implemented (BBC,2020). During the lockdown period, police recorded crime records suggest an uptick in crimes flagged as domestic abuse-related; however, there has been a steady increase in police-recorded domestic abuse-related offences in recent years as police have increased their recording of these offences.

Pandemics and natural disasters alike influence domestic and sexual assault rates. However, there has been no study into the causes of those consequences and, most specifically, how they come to affect the anticipated economic and social crisis. According to the NewYork times (2020), “Because of Covid-19 causing most countries to impose some restrictions on physical distance, domestic violence will increase, which is consistent with the understanding that domestic violence goes up whenever families spend more time together, such as the Christmas and summer vacations”. There is a direct link, as defined by sociology and gender literature, which suggests that as families spend more time together, violence may arise for no apparent reason—a dynamic that can be due to human psychology (Booth 2017; Nofziger and Kurtz 2005). Hence the lockdown, on the other hand, maybe regarded as an indirect source of crime.

In the case study area Kent, on evaluation of the data obtained from the Kent police database, it was noted that the figures put the crime committed in the county four places higher than London, which from the 12 months to June had 99 crimes per 1,000. Despite being seventh on the national list, ONS figures showed the county had the second-highest level of violent crimes in the UK and was the most crime-ridden region in the southeast. As the county fights county lines street cartels, it was also the only southern city to make the top ten of the national chart. As a result, it is not surprising that there was an increase in crime and sexual abuse offences prior to and after lockout periods. The data obtained showed that prior to the lockdown period, the highest record of sexual offence related cases in Kent was in January 2020, with a total of 6365 cases. During the lockdown and the first relation of the lockdown, the highest cases of sexual-related cases in Kent was recorded with 8103 violence and sexual offence related cases.When Covid-19 contamination spreads, different governments are required to impose limits on some facets of movement—a policy that has resulted in layoffs, work losses, and income loss. Scholars believe that low income is linked to a rise in domestic abuse (Peprah and Koomson, 2017). According to Benson and Fox (2003), couples with elevated stress levels have a 3.5 times higher risk of aggression than people with low-stress levels. According to recent studies, lockout situations triggered by rare incidents such as natural disasters and pandemics keep families in close proximity for extended periods of time, reducing the healthy time available (i.e., as a result of one party going to work), which raises the likelihood of domestic or other associated abuse happening (Uken, 2020).
